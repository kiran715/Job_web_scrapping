```python
from bs4 import BeautifulSoup
import requests
import lxml
import csv
import codecs
```

```python
with open("jobsdata.csv","w",encoding='UTF-8') as file:
    writer = csv.writer(file)
    writer.writerow(["job name","company_name","location","package","experience"])
    for i in range(1,100+1):
        url = "https://internshala.com/jobs/page-{}".format(i)
        html = requests.get(url,"lxml").text
        soup = BeautifulSoup(html)
        divide = soup.find_all("div",class_="container-fluid individual_internship visibilityTrackerItem")
#print(divide1)
        for div1 in divide:
            job = div1.find("a",class_="view_detail_button").text
            company_name = div1.find("a",class_="link_display_like_text view_detail_button").text
            location = div1.find("a",class_="location_link view_detail_button").text
            package = div1.find("div",class_="item_body salary").text
            experience = div1.find("div",class_="item_body desktop-text").text
    #print(job,company_name,location,package,experience)
            writer.writerow([job,company_name,location,package,experience])
    file.close()
```
